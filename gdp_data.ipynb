# Remove Extra Spaces

```python
df.columns = df.columns.str.strip()
```

```python
df["gdp_value"].head()
```

# changing data type of the gdp_value object to float

```python
df["gdp_value"] = df["gdp_value"].replace(r'[^0-9.]', '',regex = True)
```

```python
df["gdp_value"] = df["gdp_value"].astype(float)
```

```python
df.info()
```

# Remove Duplicate()

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df.isna().sum()
```

```python
df["gdp_value"].mean()
```

```python
df["gdp_value"] = df["gdp_value"].fillna(df["gdp_value"].mean())
```

```python
df.isna().sum()
```

# Fixing formating issue

```python
df["Country"].head(10)
```

```python
df["Country"] = df["Country"].str.strip()
```

```python
df["Country"].head(10)
```

```python
df["Country"] = df["Country"].str.lower()
```

```python
df["Country"].unique()
```

```python
df["Country"].head(10)
```

```python
df["Country"] = df["Country"].replace({
    "france":"France",
    "india":"India",
    "germany":"Germany",
    "china":"China",
    "egypt, arab rep.":"Egypt",
    "usa":"Usa",
    "hong kong sar, china":"Hong Kong",
    "south africa":"South Africa",
    "brazil":"Brazil",
    "nigeria":"Nigeria"
})
```
